```python
import pandas as pd
train = pd.read_csv("train1.csv")
```

```python
train.head()
```

```python
good_columns = [c for c in train if c not in ['Unnamed', 'Survived', 'Name', 'Ticket', 'Age', 'SibSp', 'Parch', 'Fare', 'PassengerId']]
target = "Survived"
good_columns
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

for column in good_columns:
    if train[column].dtype == 'object':
        le.fit(train[column])
        train[column] = le.transform(train[column])
```

```python
train[good_columns]
```

```python
import numpy as np
from sklearn.model_selection import train_test_split

trainT, trainTest = train_test_split(train, test_size = 0.2)
```

```python
len(trainT)
trainT
```

```python
len(trainTest)
```

```python
len(train)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100)
model.fit(trainT[good_columns], trainT[target])
```

```python
predict = model.predict(trainTest[good_columns])
```

```python
predict
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt

rms = sqrt(mean_squared_error(trainTest['Survived'], predict))
rms
```

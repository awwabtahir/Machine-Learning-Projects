```python
import pandas as pd
train = pd.read_csv("train1.csv")
test = pd.read_csv("test1.csv")
list(train)
list(test)
```

```python
list(train)
```

```python
from sklearn import svm
model = svm.SVC()
```

```python
good_columns = [c for c in train if c not in ['Unnamed', 'Survived', 'Name']]
target = "Survived"

from sklearn import preprocessing
le = preprocessing.LabelEncoder()

for column in good_columns:
    if train[column].dtype == 'object':
        le.fit(train[column])
        train[column] = le.transform(train[column])


train[good_columns].dtypes
```

```python
model.fit(train[good_columns], train[target])
```

```python
good_columns = [c for c in test if c not in ['Unnamed', 'Survived', 'Name']]
target = "Survived"

from sklearn import preprocessing
le = preprocessing.LabelEncoder()

for column in good_columns:
    if test[column].dtype == 'object':
        le.fit(test[column])
        test[column] = le.transform(test[column])


test[good_columns].dtypes
```

```python
test[target] = model.predict(test[good_columns])
```

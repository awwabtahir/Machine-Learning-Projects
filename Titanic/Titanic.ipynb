```python
smales = df.loc[(df['Sex'] == 'male') & df['Survived'].isin([1])]
nmales = df.loc[(df['Sex'] == 'male') & df['Survived'].isin([0])]
females = df.loc[df['Sex'] == 'female']
males = df.loc[df['Sex'] == 'male']
tfemales = len(females.index)
tmales = len(males.index)
sflen = len(sfemales.index)
nflen = len(nfemales.index)
smlen = len(smales.index)
nmlen = len(nmales.index)

print("Total: Female - {} || Male - {}".format(tfemales,tmales))
print("Female Survived: {} Male Survived: {}".format(sflen,smlen))
print("Female !Survived: {} Male !Survived: {}".format(nflen,nmlen))
print("{}% females survived {}% males survived".format((sflen/tfemales)*100,(smlen/tmales)*100))
```

```python
sfemales.to_csv('sfemale.csv', sep=',')
nfemales.to_csv('nfemale.csv', sep=',')
smales.to_csv('smale.csv', sep=',')
nmales.to_csv('nmale.csv', sep=',')
```
